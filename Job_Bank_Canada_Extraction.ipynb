## Recollection

```python
joblist = []
divs = browser.find_elements(By.XPATH, '//a[@class="resultJobItem"]')
```

```python
load_date = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")
load_date
```

```python
for item in divs:
    load_date = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    key_word = 'data'
    title = item.find_element(By.CLASS_NAME, "noctitle").text.strip().replace('\nVerified',' ')
    company = item.find_element(By.CLASS_NAME, "business").text.strip()
    distance = item.find_element(By.XPATH, '//span[@class="distance"]').text.strip()
    location = item.find_element(By.XPATH, '//li[@class="location"]').text.strip().replace('Location\n',' ')
    post_date = item.find_element(By.CLASS_NAME, "date").text.strip()
    salary = item.find_element(By.CLASS_NAME, "salary").text.replace('\n',' ').replace('\t',' ').replace('Salary:',' ').replace('Salary',' ').replace('$',' ').strip()
    source = item.find_element(By.CLASS_NAME, "wb-inv").text.strip()
#     telework = item.find_element(By.CLASS_NAME, "telework").text.strip()    
    link = item.get_attribute("href")
 
    job = {'load_date': load_date,
           'key_word': key_word,
           'title': title,
           'company': company,
           'distance': distance,
           'location' :location,
           'source': source,
           'post_date' : post_date,
           'salary' : salary,
#            'telework' : telework,
           'link' : link
          }
    joblist.append(job)
    
df = pd.DataFrame(joblist)
df
```

```python
df.drop_duplicates(subset ="link",
                     keep = False,
                   inplace = True
                  )
df
```

```python
joblistx = []
links_list = df['link'].values

for link in links_list:
    browser.get(link)
    link = link
    try:
        browser.find_element(By.XPATH, './/button[@id="applynowbutton"]').send_keys(Keys.RETURN)
        time.sleep(random.uniform(1.5,2.5))
    except:
        time.sleep(random.uniform(1,2))

    try:
        send_to = browser.find_element(By.XPATH, './/div[@id="howtoapply"]/p/a').text
    except:
        send_to = 'Not available'      
        
    try:
        job_number = browser.find_element(By.XPATH, './/span[@class="source-image"]/following-sibling::span[1]').text
    except:
        job_number = 'Not available'
        
    try:
        deadline = browser.find_element(By.XPATH, './/p[@property="validThrough"]').text
    except:
        deadline = 'Not available'
    
    try:
        start_date = browser.find_element(By.XPATH, './/span[@class="fa fa-calendar"]/following-sibling::span[2]').text
    except:
        start_date = 'Not available'

    try:
        NOC = browser.find_element(By.XPATH, './/span[@class="noc-no"]').text.replace('NOC',' ')+" - "+browser.find_element(By.XPATH, './/span[@class="noc-title"]').text
    except:
        NOC = 'Not available'

    try:
        job_type = browser.find_element(By.XPATH, './/span[@property="employmentType"]').text.replace('\n',' ')
    except:
        job_type = 'Not available'

    try:
        job_shift = browser.find_element(By.XPATH, './/span[@property="specialCommitments"]').text.replace('\n',' ')
    except:
        job_shift = 'Not available'
        
    try:
        job_vacancies = browser.find_element(By.XPATH, './/span[@class="fa fa-user"]/following-sibling::span[2]').text.replace('vacancies',' ').replace('vacancy',' ')
    except:
        job_vacancies = 'Not available'
        
    try:
        restriction = browser.find_element(By.XPATH, './/div[@class="job-audience"]/p').text.replace('Only apply to this job if:','restricted').replace('The employer accepts applications from:','not restricted')
    except:
        restriction = 'Not available'   

    time.sleep(random.uniform(2,3))
    browser.back()
    time.sleep(random.uniform(2,3))
    
    jobx = {'link' : link,
            'job_number' :job_number,
            'deadline': deadline,
            'start_date': start_date,
            'NOC': NOC,
            'job_type': job_type,
            'job_shift': job_shift,
            'job_vacancies': job_vacancies,
            'restriction': restriction,
            'send_to': send_to,
#             'hint': hint
            
          }
    joblistx.append(jobx)
dfx = pd.DataFrame(joblistx)
dfx
```

```python
df_jobbank= pd.merge(df, dfx, on=["link", "link"])
df_jobbank
```
